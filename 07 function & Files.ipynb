# LAMBDA EXPRATIONS

    1) Lambda function is a way to create smaall anonymous              function(withoutname)
    
    2) Lambda function are mainly used with filter() , map() , reduce()
    
    3) It can take any number or arguments , but can oonly have one expression

```python
MULTI = lambda x,y:x*y
```

```python
print(MULTI(5,8))
```

```python
# summation operatin using map() function

def summation(a,b):
    return a+b
```

```python
a = [1,3,5,6,9]
b = [1,7,9,12,7]
```

```python
c= list(map(summation, a,b))
print(c)
```

```python
my_list = [10,30,44,65,40,18,46,70,34,60]


out = list(filter(lambda x:(x%2 == 0 ), my_list))

print(out)
```

```python
out_1 = list(filter(lambda x:(x>=21), my_list))

print(out_1)
```

```python
out_2 = list(filter(lambda x:(x>=21 and x<=60) , my_list))

print(out_2)
```
